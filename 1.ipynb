```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('us_customer_data 2.csv')
```

```python
df.shape
```

```python
df.drop_duplicates(subset=['customer_id'])
```

```python
df['registration_date'] = pd.to_datetime(df['registration_date'], errors='coerce')
```

```python
df['phone']=df['phone'].str.split('x',n=1).str[0]
```

```python
df['phone']=df['phone'].str.lstrip('0')
```

```python
df['phone']=df['phone'].str.replace(r'\D','',regex=True)
```

```python
df.loc[df['phone'].str.len()>10,'phone']=df.loc[df['phone'].str.len()>10,'phone'].str[1:]
```

```python
df.loc[df['phone'].str.len()<10,'phone']=df.replace(df.loc[df['phone'].str.len()<10,'phone'],'0000000000')
```

```python
df['phone']=df['phone'].fillna('0000000000')
```

```python
df['phone']='+1 '+df['phone'].str[0:3]+'-'+df['phone'].str[3:6]+'-'+df['phone'].str[6:]
```

```python
df['phone']=df['phone'].replace('+1 000-000-0000',pd.NA) 
```

```python
df=df.fillna('NA')
```

3. Filtering high-value transactions from a CSV file for targeted promotions.
Tasks involved: Filtering, Cleaning

```python
df2=pd.read_csv('transaction_data 1.csv')
```

```python
df2
```

```python
merged_df=pd.merge(df2,df,on='customer_id',how='left')
```

```python
merged_df.dropna(subset=['amount'], inplace=True)
```

```python
merged_df
```

```python
merged_df=merged_df.groupby(['customer_id','name','email','phone','address','registration_date','loyalty_status']).agg(total_amount=('amount', 'sum')).reset_index()

```

```python
merged_df
```

```python
filtered_df=merged_df.loc[merged_df['total_amount']>1000]
```
